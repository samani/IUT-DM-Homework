```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.style.use('ggplot')
plt.figure(figsize=(14,20))

def outlier(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3 - q1

    boundLow = q1 - (1.5 * iqr)
    boundHigh = q3 + (1.5 * iqr)

    return boundLow, boundHigh


df = pd.read_csv('smartphone.csv', sep=',')
dfNumerical = df.iloc[:, [2, 3, 4, 6]]

for i in dfNumerical.columns:
    print('Column {} Outliers:'.format(i))
    col = dfNumerical[i]
    lowOut, highOut = outlier(col)
    for x in col:
        if (x < lowOut):
            print('low outlier {}'.format(x))
        elif (x > highOut):
            print('high outlier {}'.format(x))

bplotWeight = sns.boxplot(data=dfNumerical, width=0.4, palette="colorblind")
```
