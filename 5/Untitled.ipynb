```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.style.use('ggplot')

df = pd.read_csv('smartphone.csv', sep=',')
# print(df.head(10))
dfNumerical = df.iloc[:, [2, 3, 4, 6]]

normalized_df = preprocessing.normalize(dfNumerical)
normalized_df = pd.DataFrame(
    normalized_df, columns=["Capacity", "Ram", "Weight", "inch"])
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 6))
ax1.set_title('Before Normalize')
sns.kdeplot(dfNumerical['Capacity'], ax=ax1)
sns.kdeplot(dfNumerical['Ram'], ax=ax1)
sns.kdeplot(dfNumerical['Weight'], ax=ax1)
sns.kdeplot(dfNumerical['inch'], ax=ax1)
ax2.set_title('After Min-Max Normalize')
sns.kdeplot(normalized_df['Capacity'], ax=ax2)
sns.kdeplot(normalized_df['Ram'], ax=ax2)
sns.kdeplot(normalized_df['Weight'], ax=ax2)
sns.kdeplot(normalized_df['inch'], ax=ax2)
plt.show()


standardized_df = preprocessing.scale(dfNumerical)
standardized_df = pd.DataFrame(
    standardized_df, columns=["Capacity", "Ram", "Weight", "inch"])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 6))
ax1.set_title('Before Standardize')
sns.kdeplot(dfNumerical['Capacity'], ax=ax1)
sns.kdeplot(dfNumerical['Ram'], ax=ax1)
sns.kdeplot(dfNumerical['Weight'], ax=ax1)
sns.kdeplot(dfNumerical['inch'], ax=ax1)
ax2.set_title('After Standardize')
sns.kdeplot(standardized_df['Capacity'], ax=ax2)
sns.kdeplot(standardized_df['Ram'], ax=ax2)
sns.kdeplot(standardized_df['Weight'], ax=ax2)
sns.kdeplot(standardized_df['inch'], ax=ax2)
plt.show()


scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(dfNumerical)
scaled_df = pd.DataFrame(
    scaled_df, columns=["Capacity", "Ram", "Weight", "inch"])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 6))
ax1.set_title('Before Scaling')
sns.kdeplot(dfNumerical['Capacity'], ax=ax1)
sns.kdeplot(dfNumerical['Ram'], ax=ax1)
sns.kdeplot(dfNumerical['Weight'], ax=ax1)
sns.kdeplot(dfNumerical['inch'], ax=ax1)
ax2.set_title('After Min-Max Scaling')
sns.kdeplot(scaled_df['Capacity'], ax=ax2)
sns.kdeplot(scaled_df['Ram'], ax=ax2)
sns.kdeplot(scaled_df['Weight'], ax=ax2)
sns.kdeplot(scaled_df['inch'], ax=ax2)
plt.show()
```
